# Plot a Sine Wave

A minimal notebook that generates and plots a sine wave using pandas and matplotlib.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Build the data: one full period sampled at 500 points
x = np.linspace(0, 2 * np.pi, 500)
df = pd.DataFrame({"x": x, "y": np.sin(x)})
df.head()
```

```python
# Plot the sine wave
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(df["x"], df["y"], color="tab:blue", linewidth=2)
ax.set_title("Sine Wave")
ax.set_xlabel("x (radians)")
ax.set_ylabel("sin(x)")
ax.axhline(0, color="gray", linewidth=0.5)
ax.grid(True, alpha=0.3)
plt.show()
```
